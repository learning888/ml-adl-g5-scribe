```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from helpers import preprocess_data

plt.style.use('ggplot')
```

```python
train_test = preprocess_data(pd.read_csv('churn_train.csv').drop(columns='Unnamed: 0'))
```

```python
test_test = preprocess_data(pd.read_csv('churn_test.csv').drop(columns='Unnamed: 0'))
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = train_test.pop('churn_yes')

X_train, X_test, y_train, y_test = train_test_split(train_test,
                                                   y, test_size=.33,
                                                   random_state=7)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import RMSprop
from keras.callbacks import TensorBoard
```

```python
n_neurons = X_train.shape[1] // 2
input_dim = X_train.shape[1]
```

```python
tensorboards_callback_1 = TensorBoard(log_dir='./Demo/modelo_1')
```

```python
model_1 = Sequential()
model_1.add(Dense(n_neurons, input_shape=(input_dim,),
                  activation='relu', use_bias=True,
                 name= "gato"))
model_1.add(Dense(1, activation='sigmoid', name = 'out_capa'))

model_1.compile(optimizer=RMSprop(),
               loss='binary_crossentropy',
               metrics=['accuracy'])

model_1.fit(X_train, y_train, epochs=500, batch_size=512, verbose=1,
            callbacks=[tensorboards_callback_1],
           validation_data = (X_test, y_test))
```

```python
from keras.optimizers import Adam
tensorboards_callback_2 = TensorBoard(log_dir='./Demo/modelo_2')
```

```python
model_2 = Sequential()
model_2.add(Dense(n_neurons, input_shape=(input_dim, ),
                  activation='relu', use_bias=True,
                 name= "gato"))
model_2.add(Dense(n_neurons // 2,activation='relu', use_bias=True,
                 name= "gato_2"))
model_2.add(Dense(1, activation='sigmoid', name = 'out_capa'))

model_2.compile(optimizer=Adam(),
               loss='binary_crossentropy',
               metrics=['accuracy'])

model_2.fit(X_train, y_train, epochs=500, batch_size=512, verbose=1,
            callbacks=[tensorboards_callback_2],
           validation_data = (X_test, y_test))
```

```python
tensorboards_callback_3 = TensorBoard(log_dir='./Demo/modelo_3')
```

```python
model_3 = Sequential()
model_3.add(Dense(n_neurons, input_shape=(input_dim, ),
                  activation='relu', use_bias=True,
                 name= "gato"))
model_3.add(Dense(n_neurons // 2,activation='tanh', use_bias=True,
                 name= "gato_2"))
model_3.add(Dense(n_neurons // 4,activation='tanh', use_bias=True,
                 name= "gato_3"))
model_3.add(Dense(1, activation='sigmoid', name = 'out_capa'))

model_3.compile(optimizer=Adam(lr=0.00001),
               loss='binary_crossentropy',
               metrics=['accuracy'])

model_3.fit(X_train, y_train, epochs=1000, batch_size=512, verbose=1,
            callbacks=[tensorboards_callback_3],
           validation_data = (X_test, y_test))
```
